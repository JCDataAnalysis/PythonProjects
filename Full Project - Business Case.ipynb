## Which method is better to use?

Since the goal is to use the best techniques to sell the new products effectively, we have to consider how every approach has contributed to the total sales.

From the bar chart below, we see that the Email method is the one that accounts for more products sold, followed by the Call method and the Call approach.

```python
products_per_method = df.groupby('sales_method')['nb_sold'].sum().reset_index()
sns.barplot(data=products_per_method, x='sales_method', y='nb_sold', order=sorted(df['sales_method'].unique()))
plt.title('Products Sold by Sales Method')
plt.xlabel("Sales Method")
plt.ylabel("Products Sold")
plt.show()
```

```python
total_products_sold = df['nb_sold'].sum()
df.groupby('sales_method')['nb_sold'].sum() / total_products_sold * 100
```

In contrast, if we focus on total revenue generated by each approach results differ. In this case, the Email is also the one that has contributed the most, accounting for more than half of the total revenue, but more importantly the revenue generated by the Email + Call approach is almost twice the revenue generated by the Call method.

```python
revenue_per_method = df.groupby('sales_method')['revenue'].sum().reset_index()
sns.barplot(data=revenue_per_method, x='sales_method', y='revenue', order=sorted(df['sales_method'].unique()))
plt.title('Total Revenue by Sales Method')
plt.xlabel("Sales Method")
plt.ylabel("Total revenue")
plt.show()
```

```python
total_revenue = df['revenue'].sum()
df.groupby('sales_method')['revenue'].sum() / total_revenue * 100
```

The above shows that the Email + Call method tend to generate sales of more expensive products than the other methods.

Also, we notice there was a considerable difference in the revenue generated by each method over the weeks. As we see in the next line plot, total revenue generated by the Email method decreased over time, with a slight uptick after the reminder email was sent three weeks after the launch. In contrast, total revenue obtained from both the Call and Email + Call approaches increased over the first 5 weeks, with the combined approach being the most effective in boosting sales.

```python
revenue_over_time = df.groupby(['week', 'sales_method'])['revenue'].sum().reset_index()
sns.lineplot(data=revenue_over_time, x='week', y='revenue', hue='sales_method')
plt.title('Total Revenue over Time by Sales Method')
plt.xlabel("Number of weeks since the product launch")
plt.ylabel("Total Revenue")
```

Therefore, a combination of the Email and Email + Call methods would be the most effective way to increase sales and revenue consistently.

## Business Metrics

Since our goal is to use the best techniques to sell the new products effectively, I recommend that we employ two metrics to make informed decisions: **the percentage of total revenue generated by each approach in the last 6 weeks** and **the percentage of customers in each sales method in the last 6 weeks**.

Analyzing our six weeks of data, we find that more than 50% of the total revenue is attributed to the Email method, with the Call and combined approach accounting for 17% and 31%, respectively. Additionally, nearly 50% of our customers engage through the Email technique, while the Call and combined approach represent 33% and 17%.

Therefore, if the first percentage increases relative to the second, it indicates a positive sign, suggesting the method is being effective. Conversely, if the first percentage declines concerning the other, it suggests that the method might be less effective compared to the alternatives. For instance, this is the case with the Call approach, which appears to be less effective and may warrant reconsideration in our strategy.
